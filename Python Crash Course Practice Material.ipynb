```python
# Assignment 1 - user friendly approach and actual subject names
Mathematics = float(input("What's your score in Mathematics : "))
English = float(input("What's your score in English: "))
Chemistry = float(input("What's your score in Chemistry: "))
Biology = float(input("What's your score in Biology: "))
Physics = float(input("What's your score in Physics: "))

# Calculate average
total = Mathematics  + English + Chemistry + Biology + Physics
average = total / 5

# Determine grade using if, elif and else statement
if 80 <= average <= 100:
    grade = "A"
elif 60 <= average < 80:
    grade = "B"
elif 40 <= average < 60:
    grade = "C"
else:
    grade = "F"

print(f"Average Mark: {average}")
print(f"Grade: {grade}")
```

```python
# assignment 2 - a simple calculator

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))
operator = input("Enter an operator (+, -, *, /): ")

def calculate(a, b, c, operator):
    if operator == '+':
        return a + b + c
    elif operator == '-':
        return a - b - c
    elif operator == '*':
        return a * b * c
    elif operator == '/':
        if b == 0 or c == 0:
            return "Cannot divide by zero"
        return a / b / c
    else:
        return "Invalid operator"

result = calculate(a, b, c, operator)
print("Your Result is ", result)
```

```python
# assignment 3
def main():
# storing the item in the dictionary as shopping_cart
    shopping_cart = {}

    while True:
        print(" Menu ")
        print("1. Add to cart")
        print("2. View cart")
        print("3. Checkout and exit")
        
        choice = input("Choose an option (1-3): ")

        if choice == "1":
            item = input("What item are you buying from us?: ")
            try: # used this approach just in case the user inputs a non-number as the price
                price = float(input("Enter item price: $"))
                shopping_cart[item] = price
                print(f"Added '{item}' to cart at ${price:.2f}.") 
#.2f is added to display the price in exactly 2 decimal places
            except ValueError:
                print("Invalid price. Please enter a number.")
        
        elif choice == "2":
            if shopping_cart:
                print("Your shopping cart:")
                total = 0
                for item, price in shopping_cart.items():
                    print(f" {item}: ${price:.2f}")
                    total += price
                print(f"Total so far: ${total:.2f}")
            else:
                print("You haven't purchased anything.")

        elif choice == "3":
            print("\nFinal cart contents:")
            total = 0
            for item, price in shopping_cart.items():
                print(f"- {item}: ${price:.2f}")
                total += price
            print(f"Total cost: ${total:.2f}")
            print("Thank you for shopping with BlairPro... see you next time!")
            break

        else:
            print("Invalid option. Please choose from option 1, 2, or 3.")

if __name__ == "__main__":
    main()
```

```python
# assignment 4
import random

def main():
    # Generate a random number between 1 and 20
    winning_number = random.randint(1, 20)

    print("Welcome! Are you ready to guess a number?")
    print("Think of a number between 1 and 20.")

    while True:
        try:
            guess = int(input("Take a guess: "))
            
            if guess < 1 or guess > 20:
                print("Please enter a number between 1 and 20.")
                continue #this prompt enables Python to go back to the start of the while loop for the next guess
            if guess < winning_number:
                print("Too low!")
            elif guess > winning_number:
                print("Too high!")
            else:
                print(f"You won! {guess} is correct!")
                break

        except ValueError:
            print("Invalid input. Please enter a number.")

if __name__ == "__main__":
    main()
```

```python
# assignment 5
def main():
    student_names = []

    print("Enter student names. Type 'done' when finished.")

    while True:
        name = input("Enter student name: ").strip() #strip is to remove accidental spaces
        if name.lower() == 'done':
            break
        
    # Save names to a file
    with open("students.txt", "w") as file: # w indicates writing mode
        for name in student_names:
            file.write(name + "\n")

    print("\nStudent names saved to 'students.txt'. Reading back now...\n")

    # Read and display names from the file
    with open("students.txt", "r") as file: # r indicates reading mode
        for line in file:
            print(line.strip())

if __name__ == "__main__":
    main()
```

```python
13. # Python exceptions are errors that happen during execution. Python lets you catch and handle them to avoid crashing.
 # The standard way to handle potential errors in Python, like division by zero, is with a try/except block.
#try block attempts to run a piece of code that might cause an error.
#except block: If a specific error occurs, don't crash. Instead, run this other block of code.
# example
try:
    x = int(input("Enter a number: "))
    result = float(10 / x)
    print("Result is", result)
except ZeroDivisionError:
    print("You can't divide by zero!")
except ValueError:
    print("Please, enter a valid number.")
```
